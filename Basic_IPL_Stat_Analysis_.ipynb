**MAPPING THE VENUES ON THE INDIAN MAP**

```python
ball_df['venue'].unique()
```

```python
import pandas as pd

unique_venues = ball_df['venue'].unique()

venues_df = pd.DataFrame(unique_venues, columns=['venue'])

venues_df.head()
```

```python
from geopy.geocoders import Nominatim
geolocator= Nominatim(user_agent='app')
```

```python
lat_long=[]
for location in venues_df['venue']:
  location = geolocator.geocode(location)
  if location is None:
    lat_long.append(np.nan)
  else:
    geo=(location.latitude,location.longitude)
    lat_long.append(geo)
```

```python
venues_df['geo_loc']=lat_long
```

```python
venues_df
```

```python
np.array(venues_df['geo_loc'])
```

```python
import numpy as np
import pandas as pd

geo_loc_array = venues_df['geo_loc'].apply(pd.Series).values

geo_loc_array = np.where(geo_loc_array.any(axis=1)[:, None], geo_loc_array, np.nan)
geo_loc_array_no_nan = geo_loc_array[~np.isnan(geo_loc_array).any(axis=1)]
```

```python
geo_loc_array_no_nan
```

```python
lat, long = np.transpose(geo_loc_array_no_nan)

import pandas as pd

coords_df = pd.DataFrame({'latitude': lat, 'longitude': long})
```

```python
colors = {'RCB':'gold','MI':'blue',
          'CSK':'yellow','KKR':'purple',
          'GT':'dark blue','DC':'maroon',
          'RR':'pink','LSG':'sky blue',
          'SRH':'orange','PBKS':'red'}
```

```python
batsman = ball_df.groupby(['striker','batting_team'],as_index = False).agg({'match_id':'nunique','runs_off_bat':'sum'}).sort_values(by = 'runs_off_bat',ascending = False).rename(columns = {'match_id':'matches','runs_off_bat':'runs','striker':'Batsmen','batting_team':'Team'})

top_10_bat = batsman.head(10).copy(deep = True).reset_index(drop = True)

top_10_bat.index += 1

top_10_bat
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
fig = px.bar(top_10_bat,x = top_10_bat.index,y = 'runs',color = top_10_bat.Team.values,color_discrete_map = colors,text = top_10_bat.runs.values,title = 'Highest Runs in IPL 2023')

fig.update_xaxes(title_text='Batsmen', tickvals=top_10_bat.index, ticktext=top_10_bat['Batsmen'].tolist())
fig.show()
```

```python
runs_by_batter = ball_df[ball_df.wides.isna()].groupby(['striker','match_id','batting_team','innings'],as_index = False).agg({'runs_off_bat':'sum','ball':'count'}).sort_values(by = 'runs_off_bat',ascending = False).rename(columns = {'match_id':'matches','runs_off_bat':'runs','striker':'Batsmen','batting_team':'Team'})

runs_by_batter['SR'] = ((runs_by_batter['runs'] / runs_by_batter['ball']) * 100).round(2)

century = runs_by_batter[runs_by_batter.runs >= 100].reset_index(drop = True).drop(columns = ['matches'])

century.index += 1

century
```

```python
fig = px.bar(century,x = century.index,y = 'runs',color = century.Team.values,color_discrete_map = colors,text = century.runs.values,title = 'Centuries in IPL 2023')

fig.update_xaxes(title_text='Batsmen', tickvals=century.index, ticktext=century['Batsmen'].tolist())
fig.show()
```

```python
ball_df['bowler_runs'] = ball_df['runs_off_bat'] + ball_df['wides'].fillna(0) + ball_df['noballs'].fillna(0)
```

```python
run_con = ball_df.groupby(['bowler','innings','match_id','bowling_team'],as_index = False).agg({'bowler_runs':'sum'}).sort_values(by = 'bowler_runs',ascending = False).reset_index(drop = True)

run_con = pd.merge(run_con,ball_df[(ball_df.wides.isna()) & (ball_df.noballs.isna())].groupby(['bowler','innings','match_id','bowling_team'],as_index = False).agg({'ball':'count'}).sort_values(by = 'ball',ascending = False).reset_index(drop = True)[['match_id','bowler','ball']],on = ['match_id','bowler'],how = 'left')

run_con.index += 1

run_con.head(10).drop(columns =['match_id'])
```

```python
runs_bowler = run_con.head(10)

fig = px.bar(runs_bowler,x = runs_bowler.index,y = 'bowler_runs',color = runs_bowler.bowling_team.values,color_discrete_map = colors,text = runs_bowler['bowler_runs'].values,title = 'Most runs conceeded by a bowler in an innings',hover_data = ['innings'])

fig.update_xaxes(title_text='Bowler', tickvals=runs_bowler.index, ticktext=runs_bowler['bowler'].tolist())
fig.show()
```

```python
run_con['overs'] = run_con['ball'] / 6

run_con['ER'] = (run_con['bowler_runs'] / run_con['overs']).round(2)

run_con[run_con.overs == 4].sort_values(by = 'ER').head(10).reset_index(drop = True).drop(columns = ['match_id']).rename(columns = {'bowler_runs':'runs'})
```

```python
eco= run_con[run_con.overs == 4].sort_values(by = 'ER').head(10).reset_index(drop = True)

fig = px.bar(eco,x = eco.index,y = 'ER',color = eco.bowling_team.values,color_discrete_map = colors,text = eco['ER'].values,title = 'Best Econonmy Rate by a bowler in an innings',hover_data = ['innings'])

fig.update_xaxes(title_text='bowler', tickvals=eco.index, ticktext=eco['bowler'].tolist())
fig.show()
```
