```python
import csv
import ast

def classify_batsmen(batsmen_list):
    top_order = batsmen_list[:4]
    middle_order = batsmen_list[4:8]
    tailenders = batsmen_list[8:23]
    return top_order, middle_order, tailenders

def process_csv(file_path):
    team_classifications = {}

    with open(file_path, 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            team = row['batting_team']
            batsmen = ast.literal_eval(row['batsman_order'])

            top_order, middle_order, tailenders = classify_batsmen(batsmen)

            team_classifications[team] = {
                'Top Order': top_order,
                'Middle Order': middle_order,
                'Tailenders': tailenders
            }

    return team_classifications

def print_classifications(classifications):
    for team, classification in classifications.items():
        print(f"\n{team}:")
        for category, players in classification.items():
            print(f"{category}:")
            for i, player in enumerate(players, 1):
                print(f"{i}. {player}")


file_path = 'batsman_order.csv'
classifications = process_csv(file_path)
print_classifications(classifications)
```
